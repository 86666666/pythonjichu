```python
import matplotlib.pyplot as plt
import matplotlib#导入matplotlib库
# 绘制简单的折线图
matplotlib.rcParams['font.family']='SimHei'
#matplotlib本身不支持中文显示，要修改函数的参数，修改字体必须在第一行就修改
se=[1,4,9,16,25]
plt.plot(se)
plt.title('王凯旋做的图',fontsize=24)#fontsize是设置图表中文字的大小
plt.xlabel('x轴上的数据')#设置x轴上的标签和y轴上的标签

plt.ylabel('y轴上的数据')
plt.tick_params(axis='both',labelsize=14)#设置刻度的样式
plt.show()#打开matplotlib查看器
```

```python
#绘制散点图
s1=list(range(1000)) #用 range创建数字列表
s2=list(range(1000))
s3=list(range(800))
s4=list(range(800,1600))
#print(s1)
#plt.scatter(s1,s2,c='red')#散点图必须提供x,y轴的数据
#颜色映射
plt.scatter(s1,s2,c=s2,cmap=plt.cm.Blues)
plt.title('绘制散点图',fontsize=24)
plt.xlabel('x轴',fontsize=14)
plt.ylabel('y轴',fontsize=14)
plt.tick_params(axis='both',labelsize=14)
plt.axis([0,1500,0,2000])
plt.grid(True)#显示网格分隔

#自动保存列表
plt.savefig(r'C:\Users\86666\Desktop\uc下载\wenjian1.png',dpi=600)
print('下载成功')
plt.show()
```
